```python
import pandas
import duckdb
```

```python
conn=duckdb.connect()

#creating table from the initial csv file
conn.execute("""
            CREATE TABLE disease_data AS 
                SELECT * FROM read_csv_auto('initial/Disease_data.csv')
""")
```

```python
#Verifying for Nulls

null_existence = conn.execute("""
                SELECT 
                COUNT(*)
                FROM disease_data
                where              
                    Temperature IS NULL OR
                    Humidity IS NULL OR
                    "Wind Speed" IS NULL OR
                    Visibility IS NULL OR
                    Pressure IS NULL OR
                    Disease IS NULL OR
                    "Disease In number" IS NULL
""").fetchone()[0]

print(f"Null existence       : {null_existence}")
```

```python
#Verifying disease string conformity


disease_cases=conn.execute(""" SELECT DISTINCT 
                                    disease 
                                FROM disease_data """).fetchdf()

disease_cases
#outputs two indentical Disease states formatted differently !!

```

```python
#CHECKING DISEASE IN NUMBER CONFORMITY EXPECTED OUTPUT 0

disease_number_check=conn.execute("""SELECT     
                                        count(*) as UNEXPECTED_STATE 
                                    FROM disease_data 
                                    WHERE "Disease in number" not in (1,0)""").fetchdf()

disease_number_check
```

```python
#TEMPERATURE MIN MAX CHECK
temp_range=conn.execute(""" SELECT 
                                MIN(Temperature) AS min_temp,
                                MAX(Temperature) AS max_temp 
                            FROM disease_data""").fetchdf()


temp_range
```

```python
#HUMIDITY MIN MAX CHECK
humidity_range=conn.execute(""" SELECT 
                                    MIN(Humidity) AS min_humidity, 
                                    MAX(Humidity) AS max_humidity 
                                FROM disease_data""").fetchdf()

humidity_range
```

```python
#WIND SPEED MIN MAX CHECK
wind_speed_range=conn.execute("""   SELECT 
                                        MIN("Wind Speed") AS min_wind_speed, 
                                        MAX("Wind Speed") AS max_wind_speed 
                                    FROM disease_data""").fetchdf()

wind_speed_range
```

```python
#VISIBILITY MIN MAX CHECK
visibility_range=conn.execute("""SELECT  
                                    MIN("Visibility") AS min, 
                                    MAX("Visibility") AS max 
                                FROM disease_data """).fetchdf()
visibility_range
```

```python
#PRESSURE MIN MAX CHECK
pressure_range=conn.execute(""" SELECT  
                                    MIN("Pressure") AS min, 
                                    MAX("Pressure") AS max 
                                FROM disease_data""").fetchdf()
pressure_range
```

```python
#INVALID WIND BEARING CHECK, WIND BEARING SHOULD BE NO LESS THAN 0 AND NO HIGHER THAT 360
#expected output 0
invalid_wind_bearings=conn.execute("""  SELECT 
                                           count(*)  
                                        FROM disease_data
                                        WHERE "Wind Bearing"<0
                                        OR "Wind Bearing">360
                                    """).fetchdf()
invalid_wind_bearings
```
